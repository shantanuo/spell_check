```python
import pandas as pd
import numpy as np
```

```python
ndf = pd.read_csv('original_with_count.csv')
```

```python
ndf
```

```python
ndf = ndf[ndf.stem_count > 3]
ndf["count"]=ndf["count"].replace(1, np.nan) 
```

```python
g = ndf.groupby('stem')['count']
m = g.transform('count').eq(0) & g.transform('size').gt(1)
ndf = ndf[~m]
```

```python
ndf['count'] = ndf['count'].apply(str)
ndf['o_with_c'] = ndf['original'] + ' ' + ndf['count']
ndf['o_with_c1'] = np.where(ndf['original_lemma'].isnull(), ndf['o_with_c'],ndf['count'])
```

```python
ndf.groupby('stem').o_with_c1.apply(list).to_frame().to_csv('to_s.txt')
```
