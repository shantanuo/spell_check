```python
import pandas as pd

mylist = list()
for i in open("stackdata1.txt"):
    mylist.append(i.split("/"))
```

```python
df = pd.DataFrame(mylist)
df = df.replace(" ", "", regex=True)
df = df.replace("\n", "", regex=True)
df.columns = ["word", "tags"]
```

```python
def mysort(word):
    try:
        return "".join(sorted(word))
    except TypeError:
        return word
```

```python
df["tags"] = df.tags.apply(mysort)
```

```python
def mycount(word):
    try:
        return len(set(word))
    except TypeError:
        return 0
```

```python
df["tagscnt"] = df.tags.apply(mycount)
```

```python
df["tagscnt1"] = df.tags.str.len()
df["tagsdiff"] = df["tagscnt"] - df["tagscnt1"]
df["tagsdiff"] = df["tagsdiff"].fillna(0)
df.sort_values("tagsdiff")
```

```python
mys = df.groupby("tags")["word"].apply(list)
```

```python
vc = df["tags"].value_counts()
```

```python
ndf = pd.concat([vc, mys], axis=1).sort_values("tags", ascending=False)
ndf = ndf.reset_index()
ndf[["index", "word"]].to_csv("to_study.txt", index=False)
```
